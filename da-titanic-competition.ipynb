```python
import numpy as np
import pandas as pd

#The Machine learning alogorithm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

# Test train split
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv("/kaggle/input/titanic/train.csv")
train.head()
```

```python
train.describe()
```

```python
train.isnull().sum()
```

```python
train_x = train[["Pclass", "Sex"]]
train_x.head()
```

```python
train_y = train[["Survived"]]
train_y.head()
```

```python
train_x["Sex"].replace("male", 1, inplace = True)
train_x["Sex"].replace("female", 0, inplace = True)
train_x.head()
```

```python
tr_x, cv_x, tr_y, cv_y   = train_test_split(train_x, train_y, test_size = 0.30)

print(tr_x.head())
print(tr_y.head())
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(tr_x, tr_y)
```

```python
accuracy = rf.score(cv_x, cv_y)
print("Accuracy = {}%".format(accuracy * 100))
```

```python
lgr = LogisticRegression()
```

```python
lgr.fit(tr_x, tr_y)
```

```python
accuracy = lgr.score(cv_x, cv_y)
print("Accuracy = {}%".format(accuracy * 100))
```

```python
test = pd.read_csv("/kaggle/input/titanic/test.csv")
test.head()
```

```python
test_x = test[["Pclass", "Sex"]]
test_x.head()
```

```python
test_x["Sex"].replace("male", 1, inplace = True)
test_x["Sex"].replace("female", 0, inplace = True)
test_x.head()
```

```python
test_x.head()
```

```python
prd = rf.predict(test_x)
```

```python
prd
```

```python
op = test[['PassengerId']]
op['Survived'] = prd
op.head()
```
